```python
def process_age(df,cut_points,label_names):
    df["Age"] = df["Age"].fillna(-0.5)
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df
```

```python
cut_points = [-1,0,5,12,18,35,60,100]
label_names = ["Missing","Infant","Child","Teenager","Young Adult","Adult","Senior"]
df_train = process_age(df_train,cut_points,label_names)
df_test = process_age(df_test,cut_points,label_names)
```

```python
pivot = df_train.pivot_table(index="Age_categories",values='Survived')
pivot.plot.bar()
plt.show()
```

```python
def create_dummies(df,column_name):
    dummies = pd.get_dummies(df[column_name],prefix=column_name,drop_first=True)
    df = pd.concat([df,dummies],axis=1)
    return df
```

```python
for column in ["Pclass","Sex","Age_categories","Embarked","Cabin"]:
    df_train = create_dummies(df_train,column)
    df_test = create_dummies(df_test,column)
```

### Train Test Split 

```python
X_train = df_train.drop('Survived', axis=1) # data
y_train = df_train.Survived # labels
```

### Model function

```python
def Model(model,X_train, y_train):
    model.fit(X_train,y_train)
    accuracy=round(model.score(X_train, y_train) * 100, 2)
    train_pred = model_selection.cross_val_predict(model,X_train,y_train,cv=10,n_jobs = -1)
    cv_score=round(metrics.accuracy_score(y_train, train_pred) * 100, 2)
    return train_pred,accuracy,cv_score
```

### Logistic Regression

```python
LR,acc_lr,cv_lr=Model(LogisticRegression(),X_train, y_train)
print("Accuracy: %s" % acc_lr)
print("Accuracy CV 10-Fold: %s" % cv_lr)
```

### K-Nearest Neighbours 

```python
knn,acc_knn,cv_knn=Model(KNeighborsClassifier(),X_train, y_train)
print("Accuracy: %s" % acc_knn)
print("Accuracy CV 10-Fold: %s" % cv_knn)
```

### Decision Tree Classifier 

```python
dt,acc_dt,cv_dt=Model(DecisionTreeClassifier(),X_train, y_train)
print("Accuracy: %s" % acc_dt)
print("Accuracy CV 10-Fold: %s" % cv_dt)
```

### Gaussian Naive Bayes

```python
gnb,acc_gnb,cv_gnb=Model(GaussianNB(),X_train, y_train)
print("Accuracy: %s" % acc_gnb)
print("Accuracy CV 10-Fold: %s" % cv_gnb)
```

### Linear Support Vector Machines

```python
svc,acc_svc,cv_svc=Model(LinearSVC(),X_train, y_train)
print("Accuracy: %s" % acc_svc)
print("Accuracy CV 10-Fold: %s" % cv_svc)
```

### Gradient Boosting Classifier

```python
gbc,acc_gbc,cv_gbc=Model(GradientBoostingClassifier(),X_train, y_train)
print("Accuracy: %s" % acc_gbc)
print("Accuracy CV 10-Fold: %s" % cv_gbc)
```

### Random Forest Classifier 

```python
rfc,acc_rfc,cv_rfc=Model(RandomForestClassifier(),X_train, y_train)
print("Accuracy: %s" % acc_rfc)
print("Accuracy CV 10-Fold: %s" % cv_rfc)
```

### Hyperparameter Tuning Random Forest Classifier 

```python
param_grid = {"n_estimators": [5,10,23,32,42,52],
              "max_depth": [5,6,7],
              "min_samples_split": [7,8,9,10],
              "min_samples_leaf": [3,4,5],
              "max_leaf_nodes": [14,15,16]}
rfc_ht=RandomForestClassifier()
grid_search = GridSearchCV(rfc_ht, param_grid, cv=10, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_estimator_
```

```python
grid_best_gradient= RandomForestClassifier(max_depth=7, max_leaf_nodes=15, min_samples_leaf=3,
                       min_samples_split=9, n_estimators=10)
grid_best_gradient.fit(X_train,y_train)
```

```python
accuracy=round(grid_best_gradient.score(X_train, y_train) * 100, 2)
train_pred = model_selection.cross_val_predict(grid_best_gradient,X_train,y_train,cv=10,n_jobs = -1)
print("Accuracy=>",accuracy)
print("Cross Validation score=>",round(metrics.accuracy_score(y_train, train_pred) * 100, 2))
```
